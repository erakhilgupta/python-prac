```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

```python
N= 400
Data = pd.DataFrame({ 'Kilometers' : range(1, N + 1 ,1),
    'Random_Number' : np.random.uniform(0,1 , size=N)})
Data.head()
```

```python
Data['Avg_Speed_A'] = np.where((Data['Kilometers']<=50),norm.ppf(Data['Random_Number'], loc=8, scale=0.1),0)
Data['Avg_Speed_B'] = np.where((Data['Kilometers']>50) & (Data['Kilometers']<=90),norm.ppf(Data['Random_Number'], loc=16, scale=0.1),0)
Data['Avg_Speed_C'] = np.where((Data['Kilometers']>90) & (Data['Kilometers']<=390),norm.ppf(Data['Random_Number'], loc=48, scale=0.1),0)
Data['Avg_Speed_D'] = np.where((Data['Kilometers']>390),norm.ppf(Data['Random_Number'], loc=4, scale=0.1),0)
Data
```

```python
Data['Speed_per_km'] = Data.apply(lambda row: row.Avg_Speed_A + row.Avg_Speed_B + row.Avg_Speed_C + row.Avg_Speed_D, axis=1)
Data.head()
```

```python
plt.plot('Kilometers', 'Speed_per_km', data=Data)
plt.title('Distance vs Speed')
plt.xlabel('Distance (km)')
plt.ylabel('Speed (km/hr)')
plt.show()
```
